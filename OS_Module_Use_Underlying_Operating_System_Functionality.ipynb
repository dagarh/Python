```python
import os    # operating system module

# it will show you all the instances and member functions inside os module
print(dir(os) ,"\n\n")

print(os.getcwd())  # get current working directory
```

```python
# we can change the working dir 
os.chdir('/home/hdagar3/')  
print(os.getcwd())

# we can also list down the files and folders inside a directory
listOfFilesAndFolders = os.listdir()
print(listOfFilesAndFolders)


# if you wanna create a folder inside a current working directory 
os.mkdir('DataScience') # disadvantage is it won't be able to create files/folder hierarchy in one go

# os.mkdir('DataScience/BigData')  # this will give error if executed beacause at a time it will create only one file or folder

#we can use makedirs() to overcome above disadvantage
os.makedirs('Science/Chemistry/Chemicals')  # so always use this instead of above


# in order to remove these directorixes
os.rmdir('DataScience')   # used to remove only one directory at a time   (prefer this to be on safe side)
os.removedirs('Science/Chemistry/Chemicals')  # used to remove whole hierarchy 
```

```python
# renaming a file
print(os.listdir())
os.rename('Sorting_Lists_Tuples_and_Objects.ipynb','Sorting_Lists_Tuples_and_Object.ipynb')
```

```python
# this is how you can get some useful info about the file or folder.
print(os.stat('Sorting_Lists_Tuples_and_Objects.ipynb'))
print(os.stat('Sorting_Lists_Tuples_and_Objects.ipynb').st_size)  # shows size of file in bytes
print(os.stat('Sorting_Lists_Tuples_and_Objects.ipynb').st_mtime)  # last modified timestamp

tstamp = os.stat('Sorting_Lists_Tuples_and_Objects.ipynb').st_mtime
# In order to get actual time from this timestamp
from datetime import datetime    # from (module) import (class) module->collection of classes,functions,global variables
print(datetime.fromtimestamp(tstamp))   # in human readable format
```

```python
# os.walk()  ---> it yields a generator expression which is iterable. and that generator is tuple of 3 values
# first one - directory path , second one - list of directories , third one - list of files

for dirpath,dirnames,filenames in os.walk('/home/hdagar3/'):
    print('Current Path :',dirpath)
    print('Directories :',dirnames)
    print('Files : ',filenames)
    print()

# So in a tree based form it goes to each and every subdirectory and prints 3 values for all for those.
```
